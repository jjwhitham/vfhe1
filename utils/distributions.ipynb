```python
from collections import defaultdict
from matplotlib import pyplot as plt

def plot_dist(nu):
    mult_tallies = defaultdict(int)
    # count product frequencies, adding 2 for commutativity
    for i in range(-nu // 2, nu // 2):
        for j in range(i, nu // 2):
            if i * j != 0:
                mult_tallies[i * j] += 2

    # plot the distribution
    products = list(mult_tallies.keys())
    frequencies = [mult_tallies[p] for p in products]

    plt.bar(products, frequencies)
    plt.show()

plot_dist(2**6)
```
